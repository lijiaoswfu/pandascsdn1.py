```python
#SQL条件筛选
import arcpy
fc = r"C:\carl\gdb\JX.gdb\镇界"
outfc = r"C:\carl\gdb\mw.shp"
fieldname = "XZQMC"
delimfield = arcpy.AddFieldDelimiters(fc, fieldname)#为了避免识别符出错，使用这个可以
sqlWhere = delimfield + "= '马武镇'" #SQL语法
with arcpy.da.SearchCursor(fc, ["XZQMC","DCMJ"],sqlWhere) as cursor:
    for row in cursor:
        print(row[0],row[1])
        arcpy.Select_analysis(fc, outfc, sqlWhere)
        print("筛选导出完成！")
```

```python
#复制要素进入gdb，主要是多一个验证表名的程序
import arcpy
import os
arcpy.env.workspace = r"C:\carl\gdb"
outworkspace = r"C:\carl\PythonExcel\pro\Geopy\Geopy.gdb"
fclist = arcpy.ListFeatureClasses()
for shapefile in fclist:
    fcname = arcpy.Describe(shapefile).baseName
    newfcname = arcpy.ValidateTableName(fcname, outworkspace)#验证表名是否正确
    outfc = os.path.join(outworkspace, newfcname)
    arcpy.CopyFeatures_management(shapefile, outfc)
```
